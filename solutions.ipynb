```python
# Question 1: With the x and y values below, create a plot using only Matplotlib.
# You should plot y1 as a scatter plot and y2 as a line, using different colors
# and a legend.  You can name the data simply "y1" and "y2".  Make sure the
# axis labels are legible.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib notebook

#generate random data
x = pd.date_range(start='1990/1/1', end='1991/12/1', freq='MS')
y1 = np.random.normal(10, 2, len(x))
y2 = [np.sin(v)+10 for v in range(len(x))]

#figure y1 data
fig_y1, data_y1 = plt.subplots()
data_y1.scatter(x,y1,c='red',label='y1')
data_y1.legend(loc='best')
fig_y1.show()
fig_y1.savefig('q1a_plot.png')


#figure y2 data
fig_y2, data_y2 = plt.subplots()

#data_y2.plot(x,y)

data_y2.plot(x,y2,c='blue',label='y2')
data_y2.legend(loc='best')
fig_y2.show()
fig_y2.savefig('q1b_plot.png')
```

```python
# Question 3: Load the mpg.csv file that is in this repo, and create a
# plot that tests the following hypothesis: a car with an engine that has
# a higher displacement (i.e. is bigger) will get worse gas mileage than
# one that has a smaller displacement.  Test the same hypothesis for mpg
# against horsepower and weight.
```

```python
import pandas as pd

data = pd.read_csv ('mpg.csv')  
df = pd.DataFrame(data)
```

```python
import seaborn as sns


mpg_per_displacement = sns.lmplot(y='mpg',x='displacement',data=df,fit_reg=True,scatter_kws={"color": "black"}, line_kws={"color": "red"})
mpg_per_displacement.savefig("q3a_mpg_per_displacement.png")

mpg_per_horsepower = sns.lmplot(y='mpg',x='horsepower',data=df,fit_reg=True,scatter_kws={"color": "black"}, line_kws={"color": "red"})
mpg_per_horsepower.savefig("q3b_mpg_per_horsepower.png")

mpg_per_weight = sns.lmplot(y='mpg',x='weight',data=df,fit_reg=True,scatter_kws={"color": "black"}, line_kws={"color": "red"})
mpg_per_weight.savefig("q3c_mpg_per_weight.png")
```

```python
# Question 5: Continuing from question 3, create a two-by-two grid of
# subplots, where each one has mpg on the y-axis and one of displacement,
# horsepower, weight, and acceleration on the x-axis.  To clean up this 
# plot, remove the y-axis labels on the right two plots - the scale will 
# already be aligned because the mpg values are the same.
```

```python
q5b_figure, axss = plt.subplots(2, 2, sharey=True)
sns.regplot(x='displacement', y='mpg', data=df, ax=axss[0,0],scatter_kws={"color": "black"}, line_kws={"color": "red"})
sns.regplot(x='horsepower', y='mpg', data=df, ax=axss[0,1],scatter_kws={"color": "black"}, line_kws={"color": "red"})
sns.regplot(x='weight', y='mpg', data=df, ax=axss[1,0],scatter_kws={"color": "black"}, line_kws={"color": "red"})
sns.regplot(x='acceleration', y='mpg', data=df, ax=axss[1,1],scatter_kws={"color": "black"}, line_kws={"color": "red"})
for ax in axss.flat:
    ax.label_outer()
q5b_figure.savefig("q5_figure.png")
```
